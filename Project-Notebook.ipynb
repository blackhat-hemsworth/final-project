# Question 1 - Which US market is doing best with respect to real-estate? (trends)
### Sale Count
'Best' could mean a number of things, so I will try to the question from a few perspectives. To start, I will use the monthly sales by region (sales data set). I will drop some columns that will not be used, reset the columns to be datetime, set the index to be the state name, sort the data by sales in the latest month and then transpose. In the end each column will be a region and each row will be the number of sales in that region at a point in time.

```python
adjusted_sales_counts = sales.drop(labels=['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
adjusted_sales_counts = adjusted_sales_counts.iloc[1:] # filter out the aggregate row
adjusted_sales_counts.columns = pd.Index([adjusted_sales_counts.columns[0]]).union(pd.Index(pd.to_datetime(adjusted_sales_counts.columns[1:])))
adjusted_sales_counts = adjusted_sales_counts.set_index('RegionName')
adjusted_sales_counts = adjusted_sales_counts.sort_values(by=[adjusted_sales_counts.columns[-1]], ascending=False)
fig, ax = plt.subplots(figsize=(15,6))
adjusted_sales_counts.iloc[:10].T.plot(ax=ax, title="Top 10 markets by sale count - Figure 1")
adjusted_sales_counts.T.describe()
```

```python
adjusted_sales_counts.T
```

This data shows us the raw amount of sales per region. Unlike some of the other data sets, it is not smoothed or seasonaly adjusted. Because of this, you can see spikes in the summer of each year. I plotted the top 10 regions by total sales. If 'best' means total sales, then New York City is doing the best, because it had the most sales in March 2024. I think this total sales data is interesting because we can see the average and minimum unique sales per market. This could be used to establish a baseline of sales occuring in the region per month, which could help inform an investment strategy. Additionally, the seasonal trend we saw suggests that in general winter would be the best time to buy and summer the best time to sell. Another aspect worth noting, is that none of these markets have recovered to their pre-pandemic highs. But not all sales are the same, so I also want to approach this question from another angle.

### Mean Sale Price
Another way to define 'best' as a real estate market could be sale price, the best market may be the one with the highest average sale price, so that is what I will analyze here. To do this, I will use the Mean Sale Price data set and the same adjustments as I did when looking at the last data set. This data set covers a smaller time frame than that last, but a larger amount of markets. Additionally, it is smoothed and seasonally adjusted. So the outliers have been filtered out and the summertime peaks that we saw in the last data set will not be present here. 

```python
adjusted_sales_price = mean_sale_price.drop(labels=['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
adjusted_sales_price = adjusted_sales_price.iloc[1:] # filter out the aggregate row
adjusted_sales_price.columns = pd.Index([adjusted_sales_price.columns[0]]).union(pd.Index(pd.to_datetime(adjusted_sales_price.columns[1:])))
adjusted_sales_price = adjusted_sales_price.set_index('RegionName')
adjusted_sales_price = adjusted_sales_price.sort_values(by=[adjusted_sales_price.columns[-1]], ascending=False)
fig, ax = plt.subplots(figsize=(15,6))
adjusted_sales_price.iloc[:10].T.plot(ax=ax, title="Top 10 markets by average sale price - Figure 2")
plt.legend(loc='upper left')
adjusted_sales_price.T.describe()
```

This data can show us the average sale price per market across a 5 year period. The data is sorted by the highest average sale price in March 2024, the top 10 are in the plot. If 'best' is to mean highest average sale price, then Edwards CO is the best at the moment. I suspect the top locations may be vacation destinations. From glancing at the plot, I recognize some of the city names as ski resorts. That accounts for the locations with a colder climate in the top 10. The remaining locations have a warmer climate and are also popular vacation spots. San Jose is the only city in the mean sale price march 2024 top 10 to be top 100 in Size Rank. To address the broader question of real estate investment strategy, had I not seen this plot I would not have thought vacation spots would tend to sell at higher prices than the average property in a large city, so that surprised me. Also, average price seems relatively unaffected by the pandemic, unlike sale count in the last plot.

### Zillow Home Value Index (ZHVI) 
The final way I will approach this question is by opening the home value data set. This data set is similar to the last data set, except instead of containing data about average sale price, it uses Zillow's home value index measurement (ZHVI), which is an esimate based on previous sales. You can read more about it here https://www.zillow.com/research/zhvi-user-guide/. Additionally, the data is smoothed to filter out outliers and seasonally adjusted. I make the same adjustments as previosuly to format the data for a plot. 

```python
adjusted_home_values = home_values.drop(labels=['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
adjusted_home_values = adjusted_home_values.iloc[1:] # filter out the aggregate row
adjusted_home_values.columns = pd.Index([adjusted_home_values.columns[0]]).union(pd.Index(pd.to_datetime(adjusted_home_values.columns[1:])))
adjusted_home_values = adjusted_home_values.set_index('RegionName')
adjusted_home_values = adjusted_home_values.sort_values(by=[adjusted_home_values.columns[-1]], ascending=False)
fig, ax = plt.subplots(figsize=(15,6))
adjusted_home_values.iloc[:10].T.plot(ax=ax, title="Top 10 markets by ZHVI - Figure 3")
plt.legend(loc='upper left')
adjusted_home_values.T.describe()
```

This plot shows the top 10 markets by ZHVI. The top 10 in ZHVI is similar to the top 10 in average sale price, except this time San Jose comes out on top. So by this metric, San Jose is the best market in the country at the moment. 

# Question 2 - Are the sale prices and inventory listing going up or down overall in the US?
Let's take a look at both trends across the United States as a whole. 

### Sales Price and Inventory Listings -- United States


```python
mean_US_sale_price = mean_sale_price[mean_sale_price["RegionName"] == "United States"].drop(["RegionID","SizeRank","RegionType","StateName"], axis=1)
long_mean_US = mean_US_sale_price.melt(id_vars=["RegionName"], var_name = "Date", value_name = "sale_price")
long_mean_US["sale_date"] = pd.to_datetime(long_mean_US["Date"])

US_listings = for_sale_listings[for_sale_listings["RegionName"] == "United States"].drop(["RegionID","SizeRank","RegionType","StateName"], axis=1)
long_listings = US_listings.melt(id_vars=["RegionName"], var_name = "Date", value_name = "listing_count")
long_listings["sale_date"] = pd.to_datetime(long_listings["Date"])


fig, ax = plt.subplots(2,1)
sns.lineplot(data = long_mean_US,x = "sale_date", y="sale_price", ax=ax[0])
sns.lineplot(data = long_listings,x = "sale_date", y="listing_count", ax=ax[1])
_ = ax[0].title.set_text('Mean Home Sales Price (US) - Figure 4')
_ = ax[0].set(ylim=(0,500000))
_ = ax[0].set(xlim=(pd.to_datetime("01-01-2018"),pd.to_datetime("06-01-2024")))
_ = ax[1].set_title('Home Inventory (US) - Figure 5')
_ = ax[1].set(ylim=(0,1800000))
_ = ax[1].set(xlim=(pd.to_datetime("01-01-2018"),pd.to_datetime("06-01-2024")))
_ = plt.ticklabel_format(style='plain', axis='y',useOffset=False)
_ = fig.tight_layout()


```

By plotting US-wide sales price against the inventory quantity we can observe the following:
1. The mean US sales price is trending *up* between 2018 and 2024.
2. The US inventory is trending *down* over the same period.
3. The two factor seem to be correlated -- sales price increases most sharply between 2019 and mid-2022, while inventory decreases over the exact same period. An extreme "high cost" / "low inventory" moment happens around the beginning to middle of 2022.
4. The inventory trends somewhat cyclically (peaking in the middle of each year, around the summer), but the sales price is not.

# Question 3 - Should I wait to invest or should I invest now and where?
### (You could recommend more than one place)

I would not recommend investing in the current housing market -- the price has essentially plateaued between 2023 and 2024, as has the home inventory. There's a decent chance that higher interest rates cool the market. 

But in the interest of answering the question with a visual, I will assume that any current local trends in price will continue (which often happens in the housing market!).

```python
#mths = pd.date_range(start = '2018-01-01', end = '2024-01-01', freq='MS')

mean_state_sale_price = mean_sale_price.groupby(["StateName"]).mean()
mean_state_sale_price_6mo = mean_state_sale_price[["SizeRank","2023-08-31","2024-02-29"]]
mean_state_sale_price_6mo["6_month_change"] = mean_state_sale_price_6mo["2024-02-29"] - mean_state_sale_price_6mo["2023-08-31"]
mean_state_sale_price_6mo["growing?"] = mean_state_sale_price_6mo["6_month_change"] > 0

state_inventory = for_sale_listings.drop(["RegionID","SizeRank"], axis=1).groupby(["StateName"]).sum()
state_inventory_avg = state_inventory.mean(axis=1)
state_inventory_sd = state_inventory.std(axis=1)

state_inventory = state_inventory[["2024-02-29"]]
state_inventory["avg"] = state_inventory_avg
state_inventory["sd"] = state_inventory_sd
state_inventory["standardized_inventory"] = (state_inventory["2024-02-29"] - state_inventory["avg"])/state_inventory["sd"]
import seaborn.objects as so

jdf = mean_state_sale_price_6mo.merge(state_inventory, on= "StateName")
fig,ax = plt.subplots()
p = so.Plot(data=jdf,x="6_month_change",y="standardized_inventory",text="StateName").add(so.Text(halign='left'))
p.on(ax).show()
```
